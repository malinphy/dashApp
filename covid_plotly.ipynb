```
import pandas as pd 
import numpy as np 
import plotly.express as px
#!pip install geojson
#import geojson
```

```
url = 'https://raw.githubusercontent.com/malinphy/dashApp/main/CONVENIENT_global_confirmed_cases.csv'
df_covid = pd.read_csv(url)
```

```
df_covid.head()
```

```
fig = px.line(df_covid, x = df_covid["Country/Region"],y=df_covid["Germany"])
#fig = px.line(df_covid, x = "Country/Region",y="Angola")
fig.show()
```

```
numOfCases=pd.concat([df_covid["Italy"][1:], df_covid["Germany"][1:],df_covid["Spain"][1:]],axis=0)
```

```
Countries = []

for i in range(len(numOfCases)):
    if i < 302:
        Countries.append("Italy")
    elif 302<= i <604:
        Countries.append("Germany")
    else:
        Countries.append("Spain")


iso_alpha = []

for i in range(len(numOfCases)):
    if i < 302:
        iso_alpha.append("ITA")
    elif 302<= i <604:
        iso_alpha.append("DEU")
    else:
        iso_alpha.append("ESP")
        
        
c1=df_covid['Country/Region'][1:].copy()
c2=df_covid['Country/Region'][1:].copy()
c3=df_covid['Country/Region'][1:].copy()

Date = pd.concat([c1,c2,c3],axis=0)
len(Date)
```

```
Date = pd.Series(Date)
iso_alpha = pd.Series(iso_alpha)
Countries = pd.Series(Countries)
numOfCases = pd.Series(numOfCases)
```

```
list_data= list(zip(Date, iso_alpha,Countries,numOfCases))  
```

```
selected_data=pd.DataFrame(list_data,columns=["Date", "iso_alpha","Countries","numOfCases"])
selected_data.head(3)
```

```
np.max(selected_data["numOfCases"])
```

```

fig = px.choropleth(selected_data, locations="iso_alpha", 
                    color="numOfCases", hover_name="Countries", 
                    animation_frame="Date", 
                    range_color=[0,35019],  
                    color_continuous_scale='Inferno',
                    #width=1000,
                    height=800
                    )
fig.show()
```
